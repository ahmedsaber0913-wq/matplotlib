# Data Visualization with Matplotlib <br>

## 1. Introduction to Data Visualization

Data visualization is a key part of data analysis and data science, allowing us to understand complex datasets through visual representation. It helps identify patterns, trends, and outliers in data, making it easier to interpret and draw conclusions.

Some common types of plots used in data visualization are:
- Line Plots
- Bar Plots
- Scatter Plots
- Histograms

Matplotlib is a popular Python library used to create these visualizations.


<br>

## 2. What is Matplotlib?

**`Matplotlib`** is a plotting library for the Python programming language. It provides a variety of tools to generate a wide range of static, animated, and interactive plots. It's often used in conjunction with **`NumPy`** and **`Pandas`** for numerical data and data analysis.

The library is designed to be highly customizable, allowing users to create complex visualizations in a simple way.

<br>

## 3. Importing Matplotlib

Before we can start creating visualizations, we need to import the necessary libraries.<br>
The most common convention is to import **`matplotlib.pyplot`** as **`plt`**.

```python
# Importing matplotlib
import matplotlib.pyplot as plt
```

<br>

## 4. Types of Plots

### 4.1. Line Plot

A line plot is used to represent data points connected by straight lines. It's typically used to show trends over time.

```python
# Example: Line Plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### 4.2. Bar Plot

A bar plot is used to represent categorical data with rectangular bars with lengths proportional to the values they represent.

```python
# Example: Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values, color='green')
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### 4.3. Scatter Plot

A scatter plot displays values for two variables on the x and y axes. It's used to observe relationships between the variables.

```python
# Example: Scatter Plot
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### 4.4. Histogram

A histogram is used to represent the distribution of a set of continuous data. It groups data into bins and counts the number of data points in each bin.

```python
# Example: Histogram
data = np.random.randn(2000)

plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

<br>

## 5. Customizing Plots

### 5.1. Color

You can change the color of your plot elements (lines, bars, etc.) by specifying the color parameter.

```python
# Example: Customizing Color
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color='orange', marker='x')
plt.title("Customized Color Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### 5.2. Labels

You can add labels to the x and y axes using **`xlabel()`** and **`ylabel()`**.

```python
# Example: Customizing Labels
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x, y)
plt.title("Plot with Custom Labels")
plt.xlabel("X-axis: Time (s)")
plt.ylabel("Y-axis: Distance (m)")
plt.show()
```

### 5.3. Titles

The **`title()`** function allows you to set a title for your plot.

```python
# Example: Customizing Titles
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.title("Plot with Title Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### 5.4. Legends

The `legend()` function is used to add a legend to your plot, which helps to label different lines or data series.

```python
# Example: Adding Legends
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 3, 6, 10, 15]

plt.plot(x, y1, label="Series 1", color='blue')
plt.plot(x, y2, label="Series 2", color='red')
plt.title("Plot with Legends")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

## 6. Combining Pandas with Matplotlib for Visualization

Pandas integrates well with Matplotlib, allowing you to easily plot data directly from DataFrames.

```python
# Example: Pandas + Matplotlib
import pandas as pd

# Create a sample DataFrame
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [3, 7, 2, 5]}
df = pd.DataFrame(data)

# Plot the data directly from the DataFrame
df.plot(kind='bar', x='Category', y='Values', color='black')
plt.title("Pandas DataFrame Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# 7. Visualizing a dataset

```python
df = pd.read_csv('employees.csv')
```

```python
df.head()
```

## 7.1. Age Vs. Salary (Scatter Plot)

```python
# Scatter plot of Age vs Salary
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Salary'], color='purple')
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

## 7.2. Salary vs Start Date (Line Plot)

```python
# Converting "Start Date" to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])

# Plotting salary over start date
plt.figure(figsize=(10, 6))
plt.plot(df['Start Date'], df['Salary'], marker='o', color='red')
plt.title("Salary Over Time")
plt.xlabel("Start Date")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

## 7.3. Employees' Age Distribution (Histogram)

```python
# Plotting age distribution (Histogram)
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='green', edgecolor='black')
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
